# Distribution of Character Powers

```python
import pandas as pd
import seaborn as sns

%matplotlib inline

pd.set_option("display.max_rows",101)
```

```python
df = pd.read_csv('characters.csv')
```

### Distribution of Powers

```python
df.hist(bins=5, figsize=(10,8));
```

### Distribution of Equal Power Combinations

```python
feat = ['prominence','violence','diplomacy']
df.groupby(feat).size().hist(bins=6, figsize=(10,8));
```

### Power Points Across All Characters

```python
df[feat].sum().plot(kind='bar', figsize=(10,4));
```

### Characters Powers 

```python
df['score'] = df[feat].sum(axis=1)
df.sort_values(by='score',ascending=False)
```

### Distribution of Total Character Power Points

```python
df.score.hist(bins=13,figsize=(10,8));
```
