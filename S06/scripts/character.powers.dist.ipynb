# Distribution of Character Powers

```python
import pandas as pd
import seaborn as sns

%matplotlib inline

pd.set_option("display.max_rows",101)
```

```python
df = pd.read_csv('character.powers.csv')
powers = ['prominence','violence','diplomacy']
```

## Character Powers

```python
df['score'] = df[powers].sum(axis=1)
df.sort_values(by='score',ascending=False)
```

### Distribution of Powers

```python
df[powers].hist(bins=5, figsize=(10,8));
```

### Distribution of Equal Power Combinations

```python
df.groupby(powers).size().hist(bins=6, figsize=(10,8));
```

### Power Points Across All Characters

```python
df[powers].sum().plot(kind='bar', figsize=(10,4));
```

### Distribution of Total Character Power Points

```python
df.score.hist(bins=13,figsize=(10,8));
```
